```
filesRDD = sc.parallelize(pathList)
```

```
filesRDD.count()
```

```
csvFilesRDD = filesRDD.filter(lambda r: 'csv' in r)
csvFilesRDD.take(5)
```

```
csvFilesRDD.count()
```

```
import re
#Assumes input MM-DD-YYYY.csv
def convert_name(filename) :
   m = re.match("^.*([0 -9]{2})-([0-9]{2})-([0-9]{4}).csv$", filename)
   return int(m.group(3) + m.group(1) + m.group(2))
```

```
csvPairRDD = csvFilesRDD.map(lambda c:(convert_name(c),c))
csvPairRDD.take(5)
```

```
csvPairRDD = csvPairRDD.sortByKey(False)
```

```
chosenfile = csvPairRDD
```

```
chosenfile = csvPairRDD.first()[1]
```
